```python
# ============================================
# CELL 3 — SIMULATE CO2 SENSOR
# ============================================

import numpy as np

def simulate_co2_sensor(scenario="normal", duration=50):
    """
    Simulates CO2 readings over time.

    Scenarios:
    - "normal"   : baseline forest readings (~400 ppm)
    - "burning"  : charcoal burning spike (700-900 ppm)
    - "mixed"    : starts normal then spikes
    """

    time = np.arange(duration)

    if scenario == "normal":
        # Baseline forest CO2 with small natural noise
        readings = 400 + np.random.normal(0, 5, duration)

    elif scenario == "burning":
        # High CO2 — hidden charcoal burning
        readings = 750 + np.random.normal(0, 20, duration)

    elif scenario == "mixed":
        # Normal then suddenly spikes (fire starts mid-reading)
        readings = 400 + np.random.normal(0, 5, duration)
        readings[25:] = 750 + np.random.normal(0, 20, duration - 25)

    return time, readings


# --- TEST IT ---
time, readings = simulate_co2_sensor(scenario="mixed")

plt.figure(figsize=(10, 4))
plt.plot(time, readings, color="orange", linewidth=2)
plt.axhline(y=500, color="red", linestyle="--", label="Alert threshold (500ppm)")
plt.title("Simulated CO2 Sensor — Mixed Scenario")
plt.xlabel("Time (seconds)")
plt.ylabel("CO2 (ppm)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print(f"✅ CO2 sensor working | Min: {readings.min():.1f} | Max: {readings.max():.1f} ppm")
```

```python
# ============================================
# CELL 4 — DOWNLOAD REAL AUDIO SAMPLES
# ============================================

import urllib.request
import os

# We'll download free audio samples from GitHub hosted files
# These are short .wav clips representing different threats

AUDIO_SAMPLES = {
    "chainsaw"   : "https://www.soundsnap.com/files/audio/chainsaw_start.mp3",
    "fire"       : "https://upload.wikimedia.org/wikipedia/commons/8/8d/Fire_crackle.ogg",
    "footsteps"  : "https://upload.wikimedia.org/wikipedia/commons/3/3b/Footsteps_on_a_wooden_floor.ogg",
    "normal"     : "https://upload.wikimedia.org/wikipedia/commons/1/1d/Frog_in_forest.ogg"
}

os.makedirs("audio_samples", exist_ok=True)

# ---- INSTEAD: We simulate audio as numpy arrays ----
# (More reliable than downloading — no broken links)

import numpy as np
import librosa
import soundfile as sf
import matplotlib.pyplot as plt

def simulate_audio(scenario="normal", duration=3, sr=22050):
    """
    Simulates audio signals for different threat scenarios.

    Scenarios:
    - "normal"    : birds + wind (low freq, smooth)
    - "chainsaw"  : high freq aggressive noise bursts
    - "fire"      : random crackling (mid freq pops)
    - "footsteps" : rhythmic low freq thumps
    """

    t = np.linspace(0, duration, int(sr * duration))

    if scenario == "normal":
        # Soft wind + random bird-like tones
        audio = 0.1 * np.random.normal(0, 1, len(t))
        for freq in [800, 1200, 1800]:  # bird frequencies
            audio += 0.05 * np.sin(2 * np.pi * freq * t) * np.random.uniform(0.5, 1)

    elif scenario == "chainsaw":
        # Aggressive high-freq buzz + harmonics
        audio = 0.4 * np.sin(2 * np.pi * 100 * t)   # engine base
        audio += 0.3 * np.sin(2 * np.pi * 300 * t)   # blade harmonic
        audio += 0.2 * np.random.normal(0, 1, len(t)) # rough noise
        # Add burst pattern (chainsaw revs)
        burst = np.where((t % 0.5) < 0.3, 1.0, 0.3)
        audio *= burst

    elif scenario == "fire":
        # Random crackle pops
        audio = 0.05 * np.random.normal(0, 1, len(t))
        # Add random sharp pops
        pop_times = np.random.choice(len(t), size=200, replace=False)
        audio[pop_times] += np.random.uniform(0.3, 0.8, 200)

    elif scenario == "footsteps":
        # Rhythmic low frequency thumps every 0.6 seconds
        audio = 0.02 * np.random.normal(0, 1, len(t))
        step_interval = int(sr * 0.6)
        for i in range(0, len(t), step_interval):
            if i + 1000 < len(t):
                # Each step = short low freq thump
                step = np.sin(2 * np.pi * 40 * t[:1000]) * np.exp(-t[:1000] * 10)
                audio[i:i+1000] += 0.5 * step

    # Normalize
    audio = audio / (np.max(np.abs(audio)) + 1e-9)

    return audio, sr


# --- Save samples to disk (YAMNet needs .wav files) ---
os.makedirs("audio_samples", exist_ok=True)

scenarios = ["normal", "chainsaw", "fire", "footsteps"]

for scenario in scenarios:
    audio, sr = simulate_audio(scenario=scenario)
    path = f"audio_samples/{scenario}.wav"
    sf.write(path, audio, sr)
    print(f"✅ Saved: {path}")


# --- VISUALIZE ALL 4 SCENARIOS ---
fig, axes = plt.subplots(4, 2, figsize=(14, 12))

for i, scenario in enumerate(scenarios):
    audio, sr = simulate_audio(scenario=scenario)

    # Waveform
    axes[i, 0].plot(audio[:sr], linewidth=0.5)  # first 1 second
    axes[i, 0].set_title(f"{scenario.upper()} — Waveform")
    axes[i, 0].set_ylabel("Amplitude")

    # Mel Spectrogram
    mel = librosa.feature.melspectrogram(y=audio, sr=sr)
    mel_db = librosa.power_to_db(mel, ref=np.max)
    librosa.display.specshow(mel_db, sr=sr, ax=axes[i, 1], cmap="magma")
    axes[i, 1].set_title(f"{scenario.upper()} — Mel Spectrogram")

plt.tight_layout()
plt.show()

print("\n✅ Audio simulator working — 4 scenarios generated")
```

```python
# ============================================
# CELL 5 — LOAD YAMNET
# ============================================

import tensorflow_hub as hub
import tensorflow as tf
import numpy as np
import csv
import io
import urllib.request

# --- Load YAMNet model from Google's servers ---
print("⏳ Loading YAMNet... (first time takes ~30 seconds)")
yamnet_model = hub.load("https://tfhub.dev/google/yamnet/1")
print("✅ YAMNet loaded")


# --- Load the class names (521 sound labels) ---
# YAMNet comes with a label map hosted by Google
CLASS_MAP_URL = "https://raw.githubusercontent.com/tensorflow/models/master/research/audioset/yamnet/yamnet_class_map.csv"

response = urllib.request.urlopen(CLASS_MAP_URL)
lines = response.read().decode("utf-8").splitlines()

reader = csv.DictReader(lines)
CLASS_NAMES = [row["display_name"] for row in reader]

print(f"✅ Loaded {len(CLASS_NAMES)} sound class labels")
print(f"\nExample classes: {CLASS_NAMES[0]}, {CLASS_NAMES[1]}, {CLASS_NAMES[2]}...")


# --- Build prediction function ---
def yamnet_predict(wav_path, top_n=5):
    """
    Takes a .wav file path
    Returns top N predicted sound classes with scores
    """

    # Load audio
    audio, sr = librosa.load(wav_path, sr=16000, mono=True)
    # YAMNet specifically needs sr=16000

    # Run model
    scores, embeddings, spectrogram = yamnet_model(audio)

    # Average scores across time frames
    mean_scores = tf.reduce_mean(scores, axis=0).numpy()

    # Get top N results
    top_indices = np.argsort(mean_scores)[::-1][:top_n]

    results = {}
    for idx in top_indices:
        results[CLASS_NAMES[idx]] = float(mean_scores[idx])

    return results


# --- TEST IT on all 4 simulated audio samples ---
print("\n" + "="*50)
print("YAMNET PREDICTIONS ON SIMULATED AUDIO")
print("="*50)

scenarios = ["normal", "chainsaw", "fire", "footsteps"]
icons     = ["🌿",     "🪓",        "🔥",   "🐾"]

for scenario, icon in zip(scenarios, icons):
    path = f"audio_samples/{scenario}.wav"
    predictions = yamnet_predict(path)

    print(f"\n{icon} {scenario.upper()}:")
    for sound, score in predictions.items():
        bar = "█" * int(score * 30)
        print(f"   {sound:<30} {bar} {score:.3f}")
```

```python
# ============================================
# CELL 6 — ISOLATION FOREST FOR CO2 ANOMALY
# ============================================

import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# --- STEP 1: Generate Training Data ---
# We simulate many hours of "normal" forest CO2 readings
# This teaches Isolation Forest what normal looks like

np.random.seed(42)

normal_readings = 400 + np.random.normal(0, 5, 1000)  # 1000 normal samples
# This is your "training data" — all normal, no fires

print(f"✅ Training data generated")
print(f"   Normal CO2 range: {normal_readings.min():.1f} - {normal_readings.max():.1f} ppm")


# --- STEP 2: Train Isolation Forest ---
iso_forest = IsolationForest(
    contamination=0.05,   # expects 5% of data to be anomalies
    random_state=42
)

iso_forest.fit(normal_readings.reshape(-1, 1))
# reshape(-1,1) just means "one reading at a time"

print("✅ Isolation Forest trained on normal CO2 data")


# --- STEP 3: Build Prediction Function ---
def co2_anomaly_score(reading):
    """
    Takes a single CO2 reading (ppm)
    Returns:
      - is_anomaly: True/False
      - confidence: 0.0 to 1.0
      - status: human readable label
    """

    prediction = iso_forest.predict([[reading]])
    # predict returns: 1 = normal, -1 = anomaly

    raw_score = iso_forest.decision_function([[reading]])[0]
    # decision_function: more negative = more anomalous

    # Convert raw score to 0-1 confidence
    # We clip and normalize it
    confidence = float(np.clip(-raw_score * 2, 0, 1))

    is_anomaly = prediction[0] == -1

    if reading < 450:
        status = "🌿 Normal"
    elif reading < 600:
        status = "⚠️  Elevated"
    else:
        status = "🔥 Burning Detected"

    return {
        "reading"    : round(reading, 1),
        "is_anomaly" : is_anomaly,
        "confidence" : round(confidence, 3),
        "status"     : status
    }


# --- STEP 4: TEST on different scenarios ---
test_readings = {
    "Normal forest"          : 403,
    "Slightly elevated"      : 480,
    "Suspicious"             : 580,
    "Charcoal burning"       : 780,
    "Heavy burning"          : 950
}

print("\n" + "="*55)
print("ISOLATION FOREST — CO2 ANOMALY DETECTION RESULTS")
print("="*55)

for scenario, ppm in test_readings.items():
    result = co2_anomaly_score(ppm)
    bar = "█" * int(result["confidence"] * 30)
    print(f"\n{result['status']}")
    print(f"   Scenario   : {scenario}")
    print(f"   Reading    : {result['reading']} ppm")
    print(f"   Anomaly    : {result['is_anomaly']}")
    print(f"   Confidence : {bar} {result['confidence']}")


# --- STEP 5: VISUALIZE ---
# Show normal data vs anomalies on a graph

time = np.arange(80)

# Simulate a session: normal → spike → back to normal
co2_session = np.concatenate([
    400 + np.random.normal(0, 5, 30),   # normal
    650 + np.random.normal(0, 20, 20),  # burning starts
    800 + np.random.normal(0, 15, 10),  # peak burning
    400 + np.random.normal(0, 5, 20),   # back to normal
])

# Predict each reading
colors = []
for r in co2_session:
    result = co2_anomaly_score(r)
    if result["is_anomaly"]:
        colors.append("red")
    else:
        colors.append("green")

plt.figure(figsize=(12, 5))
plt.scatter(time, co2_session, c=colors, s=40, zorder=3)
plt.plot(time, co2_session, alpha=0.3, color="gray")
plt.axhline(y=450, color="orange", linestyle="--", label="Elevated threshold")
plt.axhline(y=600, color="red", linestyle="--", label="Burning threshold")
plt.scatter([], [], color="green", label="Normal")
plt.scatter([], [], color="red", label="Anomaly detected")
plt.title("CO2 Sensor — Isolation Forest Anomaly Detection")
plt.xlabel("Time (seconds)")
plt.ylabel("CO2 (ppm)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\n✅ Isolation Forest working — CO2 anomaly detection ready")
```

```python
# ============================================
# CELL 7 — FUSION LAYER + DECISION ENGINE
# ============================================

import numpy as np

# ============================================
# STEP 1 — THREAT PROFILES
# Define which sensors matter for each threat
# and how much weight each sensor carries
# ============================================

THREAT_PROFILES = {
    "illegal_logging": {
        "audio_classes" : ["Chainsaw", "Sawing", "Wood", "Cutting"],
        "co2_weight"    : 0.1,   # logging doesn't spike CO2 much
        "audio_weight"  : 0.9,   # mainly an audio threat
        "threshold"     : 0.45,
        "icon"          : "🪓"
    },
    "charcoal_burning": {
        "audio_classes" : ["Fire", "Crackle", "Burning", "Smoke"],
        "co2_weight"    : 0.6,   # burning heavily spikes CO2
        "audio_weight"  : 0.4,   # fire sound also present
        "threshold"     : 0.45,
        "icon"          : "🔥"
    },
    "animal_poaching": {
        "audio_classes" : ["Footsteps", "Gunshot", "Human voice",
                           "Speech", "Walk", "Run"],
        "co2_weight"    : 0.0,   # poaching has no CO2 signature
        "audio_weight"  : 1.0,   # purely audio threat
        "threshold"     : 0.35,  # lower threshold — rather be safe
        "icon"          : "🐾"
    }
}


# ============================================
# STEP 2 — AUDIO SCORE EXTRACTOR
# Pulls relevant scores from YAMNet output
# for a specific threat
# ============================================

def extract_audio_threat_score(yamnet_results, threat_name):
    """
    Takes YAMNet predictions dict and a threat name
    Returns highest matching score for that threat's audio classes
    """
    profile = THREAT_PROFILES[threat_name]
    relevant_classes = profile["audio_classes"]

    scores = []
    for yamnet_class, score in yamnet_results.items():
        for threat_class in relevant_classes:
            # Case insensitive partial match
            if threat_class.lower() in yamnet_class.lower():
                scores.append(score)

    if scores:
        return max(scores)   # take the highest matching score
    return 0.0


# ============================================
# STEP 3 — FUSION FUNCTION
# Combines audio + CO2 into one threat score
# per threat type
# ============================================

def fuse_sensor_data(yamnet_results, co2_result):
    """
    Takes:
      - yamnet_results : dict from yamnet_predict()
      - co2_result     : dict from co2_anomaly_score()

    Returns:
      - threat_scores  : dict of threat → final fused score
    """

    # Convert CO2 anomaly to a 0-1 score
    co2_score = co2_result["confidence"] if co2_result["is_anomaly"] else 0.0

    threat_scores = {}

    for threat_name, profile in THREAT_PROFILES.items():

        # Get audio score for this threat
        audio_score = extract_audio_threat_score(yamnet_results, threat_name)

        # Weighted fusion
        fused_score = (
            profile["audio_weight"] * audio_score +
            profile["co2_weight"]   * co2_score
        )

        threat_scores[threat_name] = {
            "fused_score"  : round(fused_score, 3),
            "audio_score"  : round(audio_score, 3),
            "co2_score"    : round(co2_score, 3),
            "threshold"    : profile["threshold"],
            "alert"        : fused_score >= profile["threshold"],
            "icon"         : profile["icon"]
        }

    return threat_scores


# ============================================
# STEP 4 — ALERT PRINTER
# Human readable output of decisions
# ============================================

def print_alerts(threat_scores, scenario_label=""):
    print("\n" + "="*55)
    print(f"  DETECTION CYCLE — {scenario_label}")
    print("="*55)

    any_alert = False

    for threat, data in threat_scores.items():
        bar_filled = int(data["fused_score"] * 30)
        bar_empty  = 30 - bar_filled
        bar = "█" * bar_filled + "░" * bar_empty

        status = "🚨 ALERT" if data["alert"] else "✅ Clear"

        print(f"""
{data['icon']}  {threat.upper().replace('_', ' ')}
   Status       : {status}
   Fused Score  : [{bar}] {data['fused_score']}
   Audio Score  : {data['audio_score']}
   CO2 Score    : {data['co2_score']}
   Threshold    : {data['threshold']}""")

        if data["alert"]:
            any_alert = True

    if not any_alert:
        print("\n  ✅ All clear — No threats detected")
    else:
        print("\n  🚨 THREAT DETECTED — Ranger alert triggered")

    print("="*55)


# ============================================
# STEP 5 — RUN FULL DETECTION CYCLES
# Test all threat scenarios end to end
# ============================================

scenarios = [
    {
        "label"    : "Peaceful forest",
        "audio"    : "normal",
        "co2_ppm"  : 403
    },
    {
        "label"    : "Illegal logging",
        "audio"    : "chainsaw",
        "co2_ppm"  : 410   # logging doesn't spike CO2
    },
    {
        "label"    : "Hidden charcoal burning",
        "audio"    : "fire",
        "co2_ppm"  : 820   # both fire sound + CO2 spike
    },
    {
        "label"    : "Animal poaching",
        "audio"    : "footsteps",
        "co2_ppm"  : 405   # no CO2 — just footsteps
    }
]

for scenario in scenarios:

    # 1. Get audio prediction from YAMNet
    audio_path     = f"audio_samples/{scenario['audio']}.wav"
    yamnet_results = yamnet_predict(audio_path)

    # 2. Get CO2 anomaly score
    co2_result     = co2_anomaly_score(scenario["co2_ppm"])

    # 3. Fuse both into threat scores
    threat_scores  = fuse_sensor_data(yamnet_results, co2_result)

    # 4. Print alerts
    print_alerts(threat_scores, scenario_label=scenario["label"])
```

```python
# ============================================
# CELL 8 — LIVE SIMULATION LOOP
# Runs continuously like a deployed system
# ============================================

import time
import random
import numpy as np
from IPython.display import clear_output

# ============================================
# STEP 1 — LIVE SENSOR READERS
# Simulate sensors sending data every N seconds
# ============================================

def live_co2_sensor(current_scenario):
    """
    Returns a live CO2 reading based on current scenario
    Adds realistic noise every reading
    """
    base = {
        "normal"   : 400,
        "burning"  : 780,
        "mixed"    : random.choice([400, 780])
    }
    reading = base.get(current_scenario, 400)
    noise   = np.random.normal(0, 8)
    return round(reading + noise, 1)


def live_audio_sensor(current_scenario):
    """
    Returns path to audio file for current scenario
    In real system this would record from microphone
    """
    audio_map = {
        "normal"    : "normal",
        "logging"   : "chainsaw",
        "burning"   : "fire",
        "poaching"  : "footsteps"
    }
    return f"audio_samples/{audio_map.get(current_scenario, 'normal')}.wav"


# ============================================
# STEP 2 — SCENARIO SCHEDULER
# Simulates a day in the forest
# Threats appear and disappear over time
# ============================================

SIMULATION_SCHEDULE = [
    {"cycle": 1,  "audio": "normal",   "co2": "normal",  "label": "Early morning — all clear"},
    {"cycle": 2,  "audio": "normal",   "co2": "normal",  "label": "Forest quiet"},
    {"cycle": 3,  "audio": "logging",  "co2": "normal",  "label": "⚠️  Chainsaw heard north sector"},
    {"cycle": 4,  "audio": "logging",  "co2": "normal",  "label": "⚠️  Chainsaw continues"},
    {"cycle": 5,  "audio": "normal",   "co2": "normal",  "label": "Sound stopped — monitoring"},
    {"cycle": 6,  "audio": "normal",   "co2": "burning", "label": "⚠️  CO2 rising east sector"},
    {"cycle": 7,  "audio": "burning",  "co2": "burning", "label": "⚠️  Fire sound + CO2 spike"},
    {"cycle": 8,  "audio": "burning",  "co2": "burning", "label": "⚠️  Burning confirmed"},
    {"cycle": 9,  "audio": "poaching", "co2": "normal",  "label": "⚠️  Footsteps detected"},
    {"cycle": 10, "audio": "normal",   "co2": "normal",  "label": "Evening — all clear"},
]


# ============================================
# STEP 3 — ALERT LOG
# Keeps track of all alerts across cycles
# ============================================

alert_log = []


def log_alert(cycle, threat, score, scenario_label):
    alert_log.append({
        "cycle"    : cycle,
        "threat"   : threat,
        "score"    : score,
        "scenario" : scenario_label,
        "time"     : time.strftime("%H:%M:%S")
    })


# ============================================
# STEP 4 — LIVE DISPLAY FUNCTION
# Prints a clean dashboard each cycle
# ============================================

def display_dashboard(cycle, scenario_label, threat_scores,
                       co2_reading, yamnet_top, alert_log):

    clear_output(wait=True)

    print("╔══════════════════════════════════════════════════════╗")
    print("║     🌿 FOREST CONSERVATION MONITORING SYSTEM 🌿      ║")
    print("╠══════════════════════════════════════════════════════╣")
    print(f"║  Cycle    : {str(cycle).ljust(42)}║")
    print(f"║  Status   : {scenario_label.ljust(42)}║")
    print(f"║  Time     : {time.strftime('%H:%M:%S').ljust(42)}║")
    print(f"║  CO2      : {str(f'{co2_reading} ppm').ljust(42)}║")
    print("╠══════════════════════════════════════════════════════╣")
    print("║  YAMNET TOP DETECTIONS                               ║")

    for sound, score in list(yamnet_top.items())[:3]:
        bar   = "█" * int(score * 20)
        line  = f"  {sound[:25]:<25} {bar:<20} {score:.2f}"
        print(f"║ {line.ljust(54)}║")

    print("╠══════════════════════════════════════════════════════╣")
    print("║  THREAT ASSESSMENT                                   ║")

    for threat, data in threat_scores.items():
        bar_f  = "█" * int(data["fused_score"] * 20)
        bar_e  = "░" * (20 - len(bar_f))
        status = "🚨 ALERT" if data["alert"] else "✅ Clear"
        line   = f"  {data['icon']} {threat[:20]:<20} [{bar_f}{bar_e}] {status}"
        print(f"║ {line.ljust(54)}║")

    print("╠══════════════════════════════════════════════════════╣")
    print("║  ALERT LOG                                           ║")

    if not alert_log:
        print("║  No alerts triggered yet                             ║")
    else:
        # Show last 4 alerts
        for entry in alert_log[-4:]:
            line = f"  [{entry['time']}] {entry['threat'][:20]:<20} score:{entry['score']:.2f}"
            print(f"║ {line.ljust(54)}║")

    print("╚══════════════════════════════════════════════════════╝")


# ============================================
# STEP 5 — RUN THE LIVE LOOP
# ============================================

print("🚀 Starting Forest Monitoring System...")
print("   Running 10 detection cycles...\n")
time.sleep(2)

for schedule in SIMULATION_SCHEDULE:

    cycle          = schedule["cycle"]
    scenario_label = schedule["label"]
    audio_scenario = schedule["audio"]
    co2_scenario   = schedule["co2"]

    # --- Read sensors ---
    co2_reading = live_co2_sensor(co2_scenario)
    audio_path  = live_audio_sensor(audio_scenario)

    # --- Run models ---
    yamnet_results = yamnet_predict(audio_path)
    co2_result     = co2_anomaly_score(co2_reading)

    # --- Fuse ---
    threat_scores  = fuse_sensor_data(yamnet_results, co2_result)

    # --- Log any alerts ---
    for threat, data in threat_scores.items():
        if data["alert"]:
            log_alert(cycle, threat, data["fused_score"], scenario_label)

    # --- Display dashboard ---
    display_dashboard(
        cycle          = cycle,
        scenario_label = scenario_label,
        threat_scores  = threat_scores,
        co2_reading    = co2_reading,
        yamnet_top     = yamnet_results,
        alert_log      = alert_log
    )

    # --- Wait before next cycle ---
    time.sleep(3)  # 3 seconds between cycles


# ============================================
# STEP 6 — FINAL SUMMARY
# ============================================

print("\n\n📋 SIMULATION COMPLETE — FULL ALERT SUMMARY")
print("="*55)

if not alert_log:
    print("No alerts were triggered during simulation")
else:
    print(f"Total alerts triggered: {len(alert_log)}\n")
    for entry in alert_log:
        print(f"  Cycle {entry['cycle']} [{entry['time']}]")
        print(f"  Threat   : {entry['threat']}")
        print(f"  Score    : {entry['score']:.3f}")
        print(f"  Scenario : {entry['scenario']}")
        print()

print("="*55)
print("✅ System shutdown complete")
```

```python
# ============================================
# CELL 9 — FULL SYSTEM VISUALIZATION
# Shows everything that happened across
# all cycles in one dashboard
# ============================================

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

# ============================================
# STEP 1 — RECONSTRUCT FULL CYCLE DATA
# Run all cycles again but store everything
# ============================================

cycle_data = []

for schedule in SIMULATION_SCHEDULE:

    cycle          = schedule["cycle"]
    audio_scenario = schedule["audio"]
    co2_scenario   = schedule["co2"]

    # Read sensors
    co2_reading    = live_co2_sensor(co2_scenario)
    audio_path     = live_audio_sensor(audio_scenario)

    # Run models
    yamnet_results = yamnet_predict(audio_path)
    co2_result     = co2_anomaly_score(co2_reading)

    # Fuse
    threat_scores  = fuse_sensor_data(yamnet_results, co2_result)

    cycle_data.append({
        "cycle"        : cycle,
        "label"        : schedule["label"],
        "co2_reading"  : co2_reading,
        "co2_anomaly"  : co2_result["is_anomaly"],
        "logging_score"  : threat_scores["illegal_logging"]["fused_score"],
        "burning_score"  : threat_scores["charcoal_burning"]["fused_score"],
        "poaching_score" : threat_scores["animal_poaching"]["fused_score"],
        "logging_alert"  : threat_scores["illegal_logging"]["alert"],
        "burning_alert"  : threat_scores["charcoal_burning"]["alert"],
        "poaching_alert" : threat_scores["animal_poaching"]["alert"],
    })

print("✅ All cycle data collected")


# ============================================
# STEP 2 — EXTRACT DATA FOR PLOTTING
# ============================================

cycles          = [d["cycle"]          for d in cycle_data]
co2_readings    = [d["co2_reading"]    for d in cycle_data]
logging_scores  = [d["logging_score"]  for d in cycle_data]
burning_scores  = [d["burning_score"]  for d in cycle_data]
poaching_scores = [d["poaching_score"] for d in cycle_data]
logging_alerts  = [d["logging_alert"]  for d in cycle_data]
burning_alerts  = [d["burning_alert"]  for d in cycle_data]
poaching_alerts = [d["poaching_alert"] for d in cycle_data]


# ============================================
# STEP 3 — BUILD THE FULL DASHBOARD FIGURE
# 4 subplots stacked vertically
# ============================================

fig, axes = plt.subplots(4, 1, figsize=(14, 16))
fig.suptitle(
    "🌿 Forest Conservation Monitoring System — Full Session Report",
    fontsize=14, fontweight="bold", y=0.98
)

# --- PLOT 1: CO2 Readings Over Time ---
ax1 = axes[0]
ax1.plot(cycles, co2_readings, color="orange", linewidth=2,
         marker="o", markersize=6, label="CO2 (ppm)")
ax1.fill_between(cycles, co2_readings, 400, alpha=0.2, color="orange")
ax1.axhline(y=500, color="red", linestyle="--",
            linewidth=1, label="Alert threshold (500ppm)")

# Mark anomaly cycles
for d in cycle_data:
    if d["co2_anomaly"]:
        ax1.axvspan(d["cycle"] - 0.4, d["cycle"] + 0.4,
                    alpha=0.15, color="red")

ax1.set_title("📊 CO2 Sensor Readings", fontweight="bold")
ax1.set_ylabel("CO2 (ppm)")
ax1.set_ylim(350, 950)
ax1.legend(loc="upper left")
ax1.grid(True, alpha=0.3)
ax1.set_xticks(cycles)


# --- PLOT 2: Threat Scores Over Time ---
ax2 = axes[1]
ax2.plot(cycles, logging_scores,  color="brown",  linewidth=2,
         marker="s", markersize=6, label="🪓 Illegal Logging")
ax2.plot(cycles, burning_scores,  color="red",    linewidth=2,
         marker="^", markersize=6, label="🔥 Charcoal Burning")
ax2.plot(cycles, poaching_scores, color="purple", linewidth=2,
         marker="D", markersize=6, label="🐾 Animal Poaching")

# Draw threshold line
ax2.axhline(y=0.45, color="gray", linestyle="--",
            linewidth=1, label="Alert threshold (0.45)")

ax2.set_title("📊 Fused Threat Scores Per Cycle", fontweight="bold")
ax2.set_ylabel("Threat Score (0-1)")
ax2.set_ylim(0, 1.0)
ax2.legend(loc="upper left")
ax2.grid(True, alpha=0.3)
ax2.set_xticks(cycles)


# --- PLOT 3: Alert Timeline ---
ax3 = axes[2]

threat_types  = ["🪓 Logging", "🔥 Burning", "🐾 Poaching"]
alert_matrix  = [logging_alerts, burning_alerts, poaching_alerts]
colors_matrix = ["brown", "red", "purple"]

for i, (threat, alerts, color) in enumerate(
    zip(threat_types, alert_matrix, colors_matrix)
):
    for c, alert in zip(cycles, alerts):
        if alert:
            ax3.scatter(c, i, s=300, color=color,
                        marker="*", zorder=5)
            ax3.axvspan(c - 0.4, c + 0.4,
                        alpha=0.1, color=color)
        else:
            ax3.scatter(c, i, s=60, color=color,
                        marker="o", alpha=0.3)

ax3.set_yticks([0, 1, 2])
ax3.set_yticklabels(threat_types)
ax3.set_title("🚨 Alert Timeline — Stars = Alert Triggered",
              fontweight="bold")
ax3.set_xlabel("Cycle")
ax3.grid(True, alpha=0.3)
ax3.set_xticks(cycles)


# --- PLOT 4: Scenario Labels Timeline ---
ax4 = axes[3]
ax4.set_xlim(0.5, len(cycles) + 0.5)
ax4.set_ylim(0, 1)
ax4.axis("off")
ax4.set_title("📋 Scenario Timeline", fontweight="bold")

for d in cycle_data:
    has_alert = (d["logging_alert"] or
                 d["burning_alert"] or
                 d["poaching_alert"])

    color = "red" if has_alert else "green"
    alpha = 0.8  if has_alert else 0.4

    ax4.add_patch(mpatches.FancyBboxPatch(
        (d["cycle"] - 0.4, 0.3), 0.8, 0.4,
        boxstyle="round,pad=0.05",
        facecolor=color, alpha=alpha,
        edgecolor="white"
    ))

    label = "🚨" if has_alert else "✅"
    ax4.text(d["cycle"], 0.5, label,
             ha="center", va="center", fontsize=12)
    ax4.text(d["cycle"], 0.15,
             f"C{d['cycle']}",
             ha="center", va="center",
             fontsize=8, color="gray")

# Legend
safe_patch  = mpatches.Patch(color="green", alpha=0.6, label="✅ All Clear")
alert_patch = mpatches.Patch(color="red",   alpha=0.8, label="🚨 Alert Triggered")
ax4.legend(handles=[safe_patch, alert_patch],
           loc="center right", fontsize=10)

plt.tight_layout()
plt.savefig("monitoring_report.png", dpi=150,
            bbox_inches="tight")
plt.show()


# ============================================
# STEP 4 — PRINT FINAL TEXT SUMMARY
# ============================================

total_alerts = sum([
    sum(logging_alerts),
    sum(burning_alerts),
    sum(poaching_alerts)
])

print("\n" + "="*55)
print("📋 FINAL SESSION SUMMARY")
print("="*55)
print(f"  Total cycles run     : {len(cycles)}")
print(f"  Total alerts fired   : {total_alerts}")
print(f"  Logging alerts       : {sum(logging_alerts)}")
print(f"  Burning alerts       : {sum(burning_alerts)}")
print(f"  Poaching alerts      : {sum(poaching_alerts)}")
print(f"\n  Report saved to      : monitoring_report.png")
print("="*55)
print("\n✅ Full system prototype complete")
print("   You have built a multi-modal conservation")
print("   monitoring system from scratch.")
```
